####  TASK 1.2: Plot different lines for categories of vehicle type and analyse the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?


<details><summary>Click here for a hint</summary>

<p>
    You will require:-
    <br>create a separate dataframe where the column recession has a value of '1'
    <br>to group the year, vehicle_type and calculate the average on the 'Automobile Sales'
    <br>one way is to -
    <br>use as_index as false else you will endup with multiple-indexed datafame
    <br>later set year as index and groupby vehicle over Sales and plot
    <br>make use of .plot() with kind = 'line'
    <br>do not forget to include labels and title
</p>
</details>


````python
# Create a new dataframe that includes only data for recession periods
df_recession = df[df['Recession'] == 1]

# Group the recession data by Year and Vehicle_Type and calculate the average sales
# Using as_index=False makes it easier to pivot the data later
df_Mline = df_recession.groupby(['Year', 'Vehicle_Type'], as_index=False)['Automobile_Sales'].sum()

# Pivot the table to have years as the index and vehicle types as columns
df_Mline_pivot = df_Mline.pivot(index='Year', columns='Vehicle_Type', values='Automobile_Sales')

# Create the plot
plt.figure(figsize=(12, 6))
df_Mline_pivot.plot(kind='line', ax=plt.gca()) # Use ax=plt.gca() to plot on the current axes

# Add titles and labels
plt.title('Sales Trend of Vehicle Types During Recessions')
plt.xlabel('Year')
plt.ylabel('Average Automobile Sales')
plt.legend(title='Vehicle Type')

# Add a grid for better readability
plt.grid(True)

# Show the plot
plt.show()

# As per instructions, you would save this plot. For example:
# plt.savefig('Line_Plot_2.png')
````

### TASK 1.3: Use the functionality of **Seaborn Library** to create a visualization to compare the sales trend per vehicle type for a recession period with a non-recession period.


<details><summary>Click here for a hint</summary>

<p>
     To visualize the average number of vehicles sold during recession and non-recession periods, you can use a bar chart
        <br> You will need to group recession average Automobile_Sales and then plot it<br>

````python
# Group the data by Recession and Vehicle_Type and calculate the average sales
# No need to create a separate dataframe, Seaborn can handle this grouping with the x, y, and hue parameters
plt.figure(figsize=(10, 6))

# Use Seaborn's barplot to create the comparison chart
# x='Recession': separates the bars into two main groups (0 and 1)
# y='Automobile_Sales': represents the height of the bars (average sales)
# hue='Vehicle_Type': creates different colored bars for each vehicle type within the recession groups
sns.barplot(x='Recession', y='Automobile_Sales', hue='Vehicle_Type', data=df)

# Customize the plot for clarity
plt.xticks(ticks=[0, 1], labels=['Non-Recession', 'Recession'])
plt.title('Average Automobile Sales by Vehicle Type: Recession vs. Non-Recession')
plt.xlabel('Period')
plt.ylabel('Average Sales')
plt.legend(title='Vehicle Type')

# Show the plot
plt.show()

# As per instructions, you would save this plot. For example:
# plt.savefig('Bar_Plot.png')
````

### TASK 1.4: Use sub plotting to compare the variations in GDP during recession and non-recession period by developing line plots for each period.
<br>Now, you want to find more insights from the data to understand the reason. <br>Plot a two line charts using subplotting to answer:-
#### How did the GDP vary over time during recession and non-recession periods? 
<br>Make use of <u>add_subplot()</u> from Matplotlib for this comparision.


````python
# Filter the data for recession and non-recession periods
recession_data = df[df['Recession'] == 1]
non_recession_data = df[df['Recession'] == 0]

# Create a figure and a set of subplots (1 row, 2 columns)
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(14, 6))

# Plot for recession periods on the first subplot (ax0)
sns.lineplot(x='Year', y='GDP', data=recession_data, ax=ax0, label='Recession')
ax0.set_title('GDP Variation During Recession Periods')
ax0.set_xlabel('Year')
ax0.set_ylabel('GDP')
ax0.grid(True)

# Plot for non-recession periods on the second subplot (ax1)
sns.lineplot(x='Year', y='GDP', data=non_recession_data, ax=ax1, label='Non-Recession')
ax1.set_title('GDP Variation During Non-Recession Periods')
ax1.set_xlabel('Year')
ax1.set_ylabel('GDP')
ax1.grid(True)

# Use tight_layout to prevent titles and labels from overlapping
plt.tight_layout()

# Show the plots
plt.show()

# As per instructions, you would save this plot. For example:
# plt.savefig('Subplot_GDP.png')
````

### TASK 1.5: Develop a Bubble plot for displaying the impact of seasonality on Automobile Sales.
<br>How has seasonality impacted the sales, in which months the sales were high or low? Check it for non-recession years to understand the trend

##### Develop a Bubble plot for displaying Automobile Sales for every month and use Seasonality Weight for representing the size of each bubble<br>
Title this plot as 'Seasonality impact on Automobile Sales'


<details><summary>Click here for a hint</summary>

<p>
     You can create Bubble Chart by calling the scatter()
        <br>Pass the 'Month' and 'Automobile_Sales' to the functions as x and y and then use Seasonality weight for size parameter</p>
</details>


````python
# Filter the data for non-recession periods
non_recession_data = df[df['Recession'] == 0]

# Create the bubble plot
plt.figure(figsize=(12, 6))

# Use a scatter plot and control the size of the points (s) with Seasonality_Weight
# We multiply the Seasonality_Weight by a factor (e.g., 1000) to make the bubble sizes more distinct
sns.scatterplot(data=non_recession_data, x='Month', y='Automobile_Sales', size='Seasonality_Weight',
                hue='Seasonality_Weight', # Optional: color bubbles by weight for better viz
                sizes=(20, 1000),  # Define the min and max bubble sizes
                alpha=0.6) # Use alpha for better visibility of overlapping bubbles

# Add titles and labels
plt.title('Seasonality Impact on Automobile Sales (Non-Recession Years)')
plt.xlabel('Month')
plt.ylabel('Automobile Sales')
plt.xticks(list(range(1, 13))) # Ensure all months are shown as ticks
plt.legend(title='Seasonality Weight', bbox_to_anchor=(1.05, 1), loc='upper left')

# Show the plot
plt.grid(True)
plt.show()

# As per instructions, you would save this plot. For example:
# plt.savefig('Bubble_Plot_Seasonality.png')
````
